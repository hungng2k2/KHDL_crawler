```python
import pandas as pd
```

```python
list_csv = ['bigdataset-04-05-2023_03-19.csv',
            'bigdataset-04-05-2023_07-11.csv',
            'bigdataset-04-05-2023_07-27.csv',
            'bigdataset-04-05-2023_08-05.csv',
            'bigdataset-04-05-2023_13-09.csv',
            'bigdataset-04-05-2023_16-23.csv']
columns = ['Hãng xe', 'Dòng xe', 'Năm đăng ký', 'Số Km đã đi', 'Tình trạng', 'Loại xe', 'Dung tích xe', 'Xuất xứ', 'Chính sách bảo hành']
```

```python
def merge_csv(list_csv):
    list_dataframe = []
    for csv_file in list_csv:
        list_dataframe.append(pd.read_csv(csv_file))
    merged_dataframe = pd.concat(list_dataframe)
    merged_dataframe.reset_index(drop=True, inplace=True)
    merged_dataframe['Năm đăng ký'] = merged_dataframe['Năm đăng ký'].astype(str)
    unique_dataframe = merged_dataframe.drop_duplicates(subset=columns, keep='last')
    return unique_dataframe
```

```python
merge_csv(list_csv).to_csv('bigdataset-v1.csv', index=False)
```
